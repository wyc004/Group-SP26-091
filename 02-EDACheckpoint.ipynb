## Background and Prior Work

Air quality can be a major factor in human health. Air quality is typically measured by the Air Quality Index (AQI), which measures the levels of five pollutants: ground-level ozone, nitrogen dioxide (NO2), sulfur dioxide (SO2), carbon monoxide (CO), and aerosols/particulate matter (PM)1. One leading contributor to poor air quality in urban areas comes from the transportation sector, especially exhaust from gasoline powered cars. Cars release many air pollutants, such as nitrogen dioxides (which includes NO2), CO, PM, formaldehyde, and carbon compounds that contribute to ground level ozone formation 2. These pollutants have a significant impact on human hearts, lungs and respiratory systems, and may lead to heart disease, stroke and cancer. Because gasoline-powered vehicles are a major source of these pollutants, factors that influence driving may also influence air quality. And one potential factor influencing gasoline consumption is gasoline price, as higher costs may reduce driving and fuel use. In this project, we would like to investigate if the cost of gasoline is related to air quality.

One previous study investigated the cost of gasoline and air pollution in Brisbane, Australia from 2010-2013 3. They found that diesel costs specifically are negatively correlated with carbon monoxide and nitrogen oxide (NOx) levels, while PM levels were not related. They found that gasoline prices were not correlated with air quality, only diesel. The authors claimed that cars emit a lower proportion of PM compared to CO/NOx, and thus the effect of driving cars on PM is negligible. They also suggest that diesel engines emit a larger proportion of NOx compared to gas powered cars, and that the behavior of drivers of diesel engines, such as trucks, is more sensitive to changes in fuel price. However, another study, covering 133 countries, found that prices of both diesel and gasoline have a negative effect on PM2.5 (a type of particulate matter)4.

A study done on diesel vs. gasoline cars in Europe did find that the emissions of diesel engines are different than those of gasoline engines under normal driving conditions, but not in a simple “one type creates more emissions than the other” way5. Diesel engines were found to have less NOx and PM emissions during most periods, but significantly peaks during times of high acceleration or low speed, when fuel efficiency is low. This is related to the temperature of the catalytic converter, which transforms harmful exhaust pollutants into less environmentally damaging gases. During periods of high acceleration or low speed, the temperature of the catalytic converter is lower, decreasing its conversion efficiency.

Aside from prices of fuels, there are many other factors that influence the emissions of cars. One consideration is lag in the effect of price changes: people only refill their gas every so often, so when the gas prices change, the effect is not felt until people need to refuel their vehicles3. Another consideration is the use of alternative fuels. If people are less willing to use gasoline in their cars, they may turn to other fuels that may release more pollutants, which is opposite of the expected trend4. Additionally, gasoline isn’t used purely for powering cars4. An aspect of time and technology must also be considered. The technologies used in cars have generally led to a decrease in pollutants in the modern era, as cars become more efficient in fuel usage, catalytic converter performance improves, and electric vehicles become more popular 24. Furthermore, the effect of human behavior and traffic is important as well. If less people are driving because of high fuel prices, traffic is better and people can drive with better fuel efficiency3. This also reduces the periods of low speed and high acceleration5, decreasing emissions.

While previous studies have explored the relationship between fuel prices and air pollution, an important limitation is their lack of geographic specificity. Prior work has all focused on a broad scale, limiting their applicability to local contexts. San Diego presents a unique case due to its coastal geography, Mediterranean climate, and distinctive transportation patterns, all of which may influence how vehicle emissions accumulate and disperse. Coastal geography and ocean winds may disperse pollutants more effectively than inland cities. Mediterranean climate with mild temperatures and lower seasonal variation may influence driving behavior and atmospheric pollutant accumulation differently. Topography and cross-border traffic (proximity to the Mexico border) may create unique traffic and emissions. Therefore, the relationship between gasoline prices and AQI observed elsewhere may not generalize to San Diego.

Additionally, prior studies rely on older datasets collected before widespread improvements in vehicle efficiency and increased electric vehicle adoption, potentially limiting their relevance to present-day air quality dynamics. This will all be dressed in our projects.

### Air Quality Index data 

This dataset describes the daily air quality index measured in San Diego. AQI is a metric used by the EPA to track levels of major pollutants in the atmosphere. Values range from 0 to 500, and values below 100 are generally considered acceptable. "Good" air quality is defined as an AQI between 0 and 50, while "moderate" is between 51 and 100. A higher AQI represents a poorer air quality, and is more dangerous for more individuals. Above 100, it's recommended to limit outdoor exposure, with the threshold of exposure limit being more severe for those who are sensitive to air quality (e.g. individuals with respiratory disease). "Unhealthy for sensitive groups" covers AQI from 101 to 150, "unhealthy" ranges 151 to 200, "very unhealthy" ranges 201 to 300, and "hazardous" ranges 301 to 500. This dataset is obtained from the EPA, a federal agency.

There are two main concerns with this dataset. First, the time range of 2020-2025 is very much impacted by COVID, so there will be confounding variables associated with this time period that may make the behavior of AQI not as it should be. Second, the defining parameter by itself is used to define the AQI. The defining parameter is the pollutant with the highest concentration, and only that number is reported. For San Diego, the defining parameter is usually PM2.5 or ozone, which are less strongly associated with car emissions. So, theoretically, the level of PM2.5 could be very high while NO2 is low. In practice, though, the processes that form nitrogen oxides and sulfur oxides also involve particulate matter and ozone, so trends seen in PM should reflect those seen in nitrogen oxides and sulfur oxides. Levels for each individual pollutants are not well kept, so the AQI is the best measure for nitrogen oxide/sulfur oxide levels available.

###Price Distribution Analysis
Price Distribution Analysis and Summary Statistics
To evaluate standard central tendencies and spread characteristics, we build mathematical summary metrics alongside an explicit distribution graph. Charting price density allows us to see if local gasoline costs cluster predictably around a historical anchor or skew wildly over time.


```python
# Summary statistics metrics
print("Gasoline Price Summary Statistics ($/Gallon):")
gasoline = gasoline.rename(columns={'APUS49E74715': 'gasoline_price'})
display(gasoline['gasoline_price'].describe().to_frame().round(3))
 
# Distribution plot
plt.figure(figsize=(9, 5))
sns.histplot(data=gasoline, x='gasoline_price', kde=True, bins=20, color='teal')
 
mean_val = gasoline['gasoline_price'].mean()
median_val = gasoline['gasoline_price'].median()
 
plt.axvline(mean_val, color='crimson', linestyle='--', linewidth=2, label=f'Mean: ${mean_val:.2f}')
plt.axvline(median_val, color='darkorange', linestyle='-', linewidth=2, label=f'Median: ${median_val:.2f}')
 
plt.title('Distribution of Monthly Gasoline Prices in San Diego County', fontsize=13, fontweight='bold', pad=15)
plt.xlabel('Gasoline Price ($ per gallon)', fontsize=11)
plt.ylabel('Count of Months', fontsize=11)
plt.legend(frameon=True)
sns.despine()
plt.tight_layout()
plt.show()
```

Distribution Interpretation:
 The generated distribution exposes a clear right-skewed layout. While the baseline arithmetic mean price hovers near $4.42 per gallon, the wide margin separating the absolute historical low ($2.95) and the inflationary peak ($6.23) indicates significant volatility. Because asymmetric market spikes skew the mean upward, the median serves as a more reliable and resilient metric for forecasting typical user cost exposure.


###Chronological Trend Tracking
Sequential Time-Series Trend Tracking
Because this data relies on a precise, sequential timeline, a static distribution profile conceals how pricing metrics dynamically reacted to historical benchmarks. We map out the data points sequentially to separate standard seasonal fluctuations from intense, macroeconomic global events.


```python
# Time series plot
df_sorted = gasoline.sort_values('observation_date')
 
plt.figure(figsize=(13, 6))
plt.plot(df_sorted['observation_date'], df_sorted['gasoline_price'], color='navy', linewidth=2.5, marker='o', markersize=4)
 
# Highlight the core pandemic period noted in the data description
plt.axvspan(pd.to_datetime('2020-03-01'), pd.to_datetime('2021-12-01'),
        	color='grey', alpha=0.15, label='Core Pandemic Lockdowns / Disruption')
 
# Highlight the historic geopolitical price peak
plt.axvline(pd.to_datetime('2022-06-01'), color='crimson', linestyle=':', linewidth=2, label='2022 Energy Price Peak')
 
plt.title('Chronological Gasoline Price Trends (San Diego County)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Timeline (Years)', fontsize=11)
plt.ylabel('Gasoline Price ($ per gallon)', fontsize=11)
plt.legend(frameon=True, loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

Trend Interpretation:
 The chronological mapping tracks a highly turbulent economic timeline containing four clear phases:
Pre-Pandemic Baseline (2018–2020): Prices hovered within a predictable operational band between $3.40 and $4.10.
Pandemic Demand Shock (Mid 2020): Public stay-at-home mandates forced an unprecedented collapse in vehicle travel, driving prices to an absolute floor of $2.95.
