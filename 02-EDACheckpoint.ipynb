```python
import matplotlib.pyplot as plt
import seaborn as sns
AQI.describe()
```

The average AQI across the dataset is approximately 71, indicating that air quality generally falls within the “Moderate” range. While air conditions are usually acceptable, certain pollutants may still pose risks for sensitive individuals.

Although many days exhibit moderate air quality, the maximum AQI value of 192 indicates the presence of severe pollution events during the observed period.

The AQI standard deviation of approximately 23 suggests substantial variation in daily air quality conditions over time.

The number of reporting monitoring sites varies considerably throughout the dataset, ranging from 1 to 15 stations per day. This may influence how representative AQI measurements are for certain time periods.

The dataset contains over 10 years of daily AQI observations, providing a strong foundation for analyzing long-term air quality trends. Average AQI levels fall within the moderate range, though substantial variability exists across the dataset. While many days exhibit relatively healthy air quality, several extreme pollution events are present, as indicated by AQI values approaching 200. Additionally, variation in the number of monitoring sites reporting each day suggests that spatial coverage may differ across time periods.

```python
# A distribution histogram to see if data is skewed/averages/spikes
plt.figure(figsize=(8,5))
sns.histplot(AQI['AQI'], bins=30)
plt.title('Distribution of AQI Values')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()
```

The AQI distribution looks moderately right-skewed, with most observations concentrated between approximately 50 and 80 AQI. This suggests that air quality is generally moderate on most days, while occasional high-pollution events create a long upper tail in the distribution. Extremely high AQI values are relatively rare but still present throughout the dataset.

```python
#time series plot to see aqi over time
#checking for trends in data
plt.figure(figsize=(14,6))
plt.plot(AQI['Date'], AQI['AQI'])
plt.title('AQI Over Time')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.show()
#add an interpretation below
```

The time-series visualization demonstrates substantial day-to-day variability in AQI levels over the ten-year period. Several major pollution spikes are visible, with AQI occasionally exceeding 150, indicating periods of unhealthy air quality. Although no strong long-term upward or downward trend is immediately apparent, recurring spikes suggest possible seasonal or environmental influences such as wildfire smoke or weather-related atmospheric conditions.

```python
#monthly averages
import calendar

AQI['Month'] = AQI['Date'].dt.month

monthly_avg = AQI.groupby('Month')['AQI'].mean()

month_names = [calendar.month_abbr[i] for i in monthly_avg.index]

plt.figure(figsize=(10,5))
plt.bar(month_names, monthly_avg)

plt.title('Average AQI by Month')
plt.xlabel('Month')
plt.ylabel('Average AQI')

plt.show()
```

Average AQI levels vary across months, indicating a seasonal pattern in air quality. AQI values tend to increase during the summer and early fall months, particularly from June through October, while lower averages are observed during late winter and early spring. This pattern may reflect seasonal environmental factors such as higher temperatures, wildfire activity, or changes in atmospheric circulation.
