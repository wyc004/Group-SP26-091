## Background and Prior Work

Air quality can be a major factor in human health. Air quality is typically measured by the Air Quality Index (AQI), which measures the levels of five pollutants: ground-level ozone, nitrogen dioxide (NO2), sulfur dioxide (SO2), carbon monoxide (CO), and aerosols/particulate matter (PM)<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). One leading contributor to poor air quality in urban areas comes from the transportation sector, especially exhaust from gasoline powered cars. Cars release many air pollutants, such as nitrogen dioxides (which includes NO2), CO, PM, formaldehyde, and carbon compounds that contribute to ground level ozone formation <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). These pollutants have a significant impact on human hearts, lungs and respiratory systems, and may lead to heart disease, stroke and cancer. Because gasoline-powered vehicles are a major source of these pollutants, factors that influence driving may also influence air quality. And one potential factor influencing gasoline consumption is gasoline price, as higher costs may reduce driving and fuel use. In this project, we would like to investigate if the cost of gasoline is related to air quality. 

One previous study investigated the cost of gasoline and air pollution in Brisbane, Australia from 2010-2013 <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). They found that diesel costs specifically are negatively correlated with carbon monoxide and nitrogen oxide (NOx) levels, while PM levels were not related. They found that gasoline prices were not correlated with air quality, only diesel. The authors claimed that cars emit a lower proportion of PM compared to CO/NOx, and thus the effect of driving cars on PM is negligible. They also suggest that diesel engines emit a larger proportion of NOx compared to gas powered cars, and that the behavior of drivers of diesel engines, such as trucks, is more sensitive to changes in fuel price. However, another study, covering 133 countries, found that prices of both diesel and gasoline have a negative effect on PM2.5 (a type of particulate matter)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). 

A study done on diesel vs. gasoline cars in Europe did find that the emissions of diesel engines are different than those of gasoline engines under normal driving conditions, but not in a simple “one type creates more emissions than the other” way<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). Diesel engines were found to have less NOx and PM emissions during most periods, but significantly peaks during times of high acceleration or low speed, when fuel efficiency is low. This is related to the temperature of the catalytic converter, which transforms harmful exhaust pollutants into less environmentally damaging gases. During periods of high acceleration or low speed, the temperature of the catalytic converter is lower, decreasing its conversion efficiency.

Aside from prices of fuels, there are many other factors that influence the emissions of cars. One consideration is lag in the effect of price changes: people only refill their gas every so often, so when the gas prices change, the effect is not felt until people need to refuel their vehicles<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Another consideration is the use of alternative fuels. If people are less willing to use gasoline in their cars, they may turn to other fuels that may release more pollutants, which is opposite of the expected trend<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Additionally, gasoline isn’t used purely for powering cars<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). An aspect of time and technology must also be considered. The technologies used in cars have generally led to a decrease in pollutants in the modern era, as cars become more efficient in fuel usage, catalytic converter performance improves, and electric vehicles become more popular <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Furthermore, the effect of human behavior and traffic is important as well. If less people are driving because of high fuel prices, traffic is better and people can drive with better fuel efficiency<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This also reduces the periods of low speed and high acceleration<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5), decreasing emissions. 

While previous studies have explored the relationship between fuel prices and air pollution, an important limitation is their lack of geographic specificity. Prior work has all focused on a broad scale, limiting their applicability to local contexts. San Diego presents a unique case due to its coastal geography, Mediterranean climate, and distinctive transportation patterns, all of which may influence how vehicle emissions accumulate and disperse. Coastal geography and ocean winds may disperse pollutants more effectively than inland cities. Mediterranean climate with mild temperatures and lower seasonal variation may influence driving behavior and atmospheric pollutant accumulation differently. Topography and cross-border traffic (proximity to the Mexico border) may create unique traffic and emissions. Therefore, the relationship between gasoline prices and AQI observed elsewhere may not generalize to San Diego. 

Additionally, prior studies rely on older datasets collected before widespread improvements in vehicle efficiency and increased electric vehicle adoption, potentially limiting their relevance to present-day air quality dynamics. This will all be dressed in our projects.

### Air Quality Index data 

This dataset describes the daily air quality index measured in San Diego. AQI is a metric used by the EPA to track levels of major pollutants in the atmosphere. Values range from 0 to 500, and values below 100 are generally considered acceptable. "Good" air quality is defined as an AQI between 0 and 50, while "moderate" is between 51 and 100. A higher AQI represents a poorer air quality, and is more dangerous for more individuals. Above 100, it's recommended to limit outdoor exposure, with the threshold of exposure limit being more severe for those who are sensitive to air quality (e.g. individuals with respiratory disease). "Unhealthy for sensitive groups" covers AQI from 101 to 150, "unhealthy" ranges 151 to 200, "very unhealthy" ranges 201 to 300, and "hazardous" ranges 301 to 500. This dataset is obtained from the EPA, a federal agency.

There are two main concerns with this dataset. First, the time range of 2020-2025 is very much impacted by COVID, so there will be confounding variables associated with this time period that may make the behavior of AQI not as it should be. Second, the defining parameter by itself is used to define the AQI. The defining parameter is the pollutant with the highest concentration, and only that number is reported. For San Diego, the defining parameter is usually PM2.5 or ozone, which are less strongly associated with car emissions. So, theoretically, the level of PM2.5 could be very high while NO2 is low. In practice, though, the processes that form nitrogen oxides and sulfur oxides also involve particulate matter and ozone, so trends seen in PM should reflect those seen in nitrogen oxides and sulfur oxides. Levels for each individual pollutants are not well kept, so the AQI is the best measure for nitrogen oxide/sulfur oxide levels available.

#Monthly Gasoline Price Dataset

This dataset contains monthly gasoline price data for San Diego County beginning on January 1, 2018, ending April 2026. Each row represents one monthly observation, and the two columns record the date and the gasoline price for that month. This dataset is useful for our project because it provides a localized measure of gasoline prices rather than a broad national or statewide average.

The most relevant variables are the date variable and the gasoline price variable. The date variable identifies the month and year of each observation. The gasoline price variable measures the average gasoline price in San Diego County, in dollars per gallon. Because the data is monthly, it can be used to examine how gasoline prices changed over time in San Diego County, including periods of increase, decrease, and fluctuation. 

One advantage of this dataset is that it is specific to San Diego County. This makes it more relevant for analyzing local transportation costs and local economic conditions than a national or state gasoline price dataset would be. Since gasoline prices can vary across regions, using county-level data gives a more accurate picture of what San Diego residents may have experienced. 

One limitation of this dataset is the effect of the COVID-19 pandemic. Since the dataset begins in 2018, it includes the pandemic period, when travel behavior, fuel demand, commuting patterns, and broader economic conditions changed dramatically. These pandemic-related disruptions may affect gasoline prices in ways that are unusual compared with normal market conditions. Therefore, any analysis using this dataset should consider that some price changes may reflect pandemic effects rather than ordinary gasoline market trends.

The raw file also contained extra rows (it has blank rows since 1990s for unknown reasons) before the actual data table, so I skipped the first few rows when loading the dataset.
```python
# Find possible outliers using the IQR method
outlier_summary = {}

for col in numeric_cols:
    Q1 = gasoline[col].quantile(0.25)
    Q3 = gasoline[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = gasoline[
        (gasoline[col] < lower_bound) | 
        (gasoline[col] > upper_bound)
    ]
    
    outlier_summary[col] = {
        "lower_bound": lower_bound,
        "upper_bound": upper_bound,
        "number_of_outliers": len(outliers)
    }

outlier_summary
```
