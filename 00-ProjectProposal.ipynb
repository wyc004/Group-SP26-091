Does the price of motor vehicle fuel affect the levels of air pollution in San Diego in the past 10 years? ## Background and Prior Work

Air quality can be a major factor in human health. Air quality is typically measured by the Air Quality Index (AQI), which measures the levels of five pollutants: ground-level ozone, nitrogen dioxide (NO2), sulfur dioxide (SO2), carbon monoxide (CO), and aerosols/particulate matter (PM)<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1). One leading contributor to poor air quality in urban areas comes from the transportation sector, especially exhaust from gasoline powered cars. Cars release many air pollutants, such as nitrogen dioxides (which includes NO2), CO, PM, formaldehyde, and carbon compounds that contribute to ground level ozone formation <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2). Most of the pollutants measured by AQI are released by cars, so it stands to reason that how much gasoline is consumed is linked to air quality. In this project, we would like to investigate if the cost of gasoline is related to air quality. 

One previous study investigated the cost of gasoline and air pollution in Brisbane, Australia from 2010-2013 <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). They found that diesel costs specifically are negatively correlated with carbon monoxide and nitrogen oxide (NOx) levels, while PM levels were not related. They found that gasoline prices were not correlated with air quality, only diesel. The authors claimed that cars emit a lower proportion of PM compared to CO/NOx, and thus the effect of driving cars on PM is negligible. They also suggest that diesel engines emit a larger proportion of NOx compared to gas powered cars, and that the behavior of drivers of diesel engines, such as trucks, is more sensitive to changes in fuel price. However, another study, covering 133 countries, found that prices of both diesel and gasoline have a negative effect on PM2.5 (a type of particulate matter)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). 

A study done on diesel vs. gasoline cars in Europe did find that the emissions of diesel engines are different than those of gasoline engines under normal driving conditions, but not in a simple “one type creates more emissions than the other” way<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5). Diesel engines were found to have less NOx and PM emissions during most periods, but significantly peaks during times of high acceleration or low speed, when fuel efficiency is low. This is related to the temperature of the catalytic converter, which transforms harmful exhaust pollutants into less environmentally damaging gases. During periods of high acceleration or low speed, the temperature of the catalytic converter is lower, decreasing its conversion efficiency.

Aside from prices of fuels, there are many other factors that influence the emissions of cars. One consideration is lag in the effect of price changes: people only refill their gas every so often, so when the gas prices change, the effect is not felt until people need to refuel their vehicles<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). Another consideration is the use of alternative fuels. If people are less willing to use gasoline in their cars, they may turn to other fuels that may release more pollutants, which is opposite of the expected trend<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Additionally, gasoline isn’t used purely for powering cars<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). An aspect of time and technology must also be considered. The technologies used in cars have generally led to a decrease in pollutants in the modern era, as cars become more efficient in fuel usage, catalytic converter performance improves, and electric vehicles become more popular <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4). Furthermore, the effect of human behavior and traffic is important as well. If less people are driving because of high fuel prices, traffic is better and people can drive with better fuel efficiency<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). This also reduces the periods of low speed and high acceleration<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5), decreasing emissions.

## Hypothesis


In case San Diego records high increases in gasoline prices, air quality in the region is bound to improve because at high fuel prices people may not feel like driving because of the higher costs and less vehicle emissions which is the major source of air pollution in the region. With the rise in historical gasoline prices in San Diego, the levels of harmful air pollutants, especially ground-level ozone (O₃) and nitrogen dioxide (NO₂), and fine particulate matter (PM2. 5) will be reduced. On the other side, when gas prices decrease it will also cause more people to use their vehicles and the quality of air will also drop. That is to say that gas prices and air pollution levels in San Diego are indirectly related: the higher the price the cleaner the air and the lower the price the dirtier the air. The proposal puts its roots in the simple economics of behavior with the established environmental science. Exposing consumers to a direct financial incentive of behavioral change when the price of gasoline increases. In the past, high fuel prices have been associated with low vehicle miles traveled (VMT), using more public transportation, carpooling, and moving to more fuel-efficient vehicles. As cars are the most common sources of air pollution in urban Southern California (they contribute a significant portion of NOℓ and volatile organic compounds (VOC) affecting smog), any real change in driving behavior must have a quantifiable effect on air quality. The case of San Diego is a place where one can specifically look at this relationship. It is a city highly reliant on cars and is underserved by a weak public transportation system compared to the scale of its space. That is, the driving behavior is particularly vulnerable to price signals of fuel, as the residents have fewer convenient options. Another factor that predisposes San Diego to smog is its warm, sunny location and geographic basin where smog can readily form and that any minimal decrease in vehicle emissions would cause discernible changes in the number of pollutants. Using historical data, the periods of high and low gasoline prices have been somewhat obvious, such as the price spikes of 2008, 20112012, and 2022, and price crashes of 2009 and 2020. These variations create natural experimental situations that result in an opportunity to test the hypothesis of whether a change in fuel prices coincides with a change in air quality measurements favored at the San Diego region by monitoring stations. The relationship between higher gas prices and higher air quality is the guiding principle of the anticipated result as the cost is one of the most influential controllers of consumers. People drive less when driving costs more, and people drive more, the atmosphere improves. This hypothesis relates economic statistics to environmental results in a manner intuitively and tactically fitting to how market dynamics can unintentionally be used to answer the question of how markets can help immensely reduce pollution.

## Data

To examine the relationship between San Diego air quality and gasoline prices from 2018 to 2026, we need the air quality data, and gasoline prices from 2018 to 2026 in San Diego.
 
For air quality, our ideal dataset would include daily or hourly measurements of ozone (O₃), nitrogen dioxide (NO₂), carbon dioxide (CO₂), sulfur dioxide (SO₂), lead (Pb), PM2.5, and PM10 collected across multiple monitoring stations throughout San Diego County. These data could be sourced from the San Diego County Air Pollution Control District (https://www.sdapcd.org/content/sdapcd/air-quality/reports-and-plans.html), which provides official annual regional air monitoring reports. Ideally we can obtain the monthly data by sending request to the website, but since the website has not responded, we now only have the yearly data of different places in San Diego. Also an ideal dataset on this question would also include relevant control variables such as temperature, wind speed, and humidity to account for other influences on air quality.
 
For gasoline prices, the ideal dataset would include daily, location-specific prices for San Diego County, categorized by fuel type (e.g., regular, mid-grade, premium). But for now, we can only obtain the monthly average price for the whole San Diego. These data could be obtained from the Federal Reserve Bank of St. Louis (https://fred.stlouisfed.org/series/APUS49E74715). We are still looking for data that can meet our requirements.
 
The difference between the ideal dataset and the real datasets brings difficulty. Gasoline price data are typically reported at monthly intervals and may be aggregated. Additionally, the absence of other control variables(weather conditions etc) may introduce confounding effects, limiting the ability for us to isolate the impact of gasoline prices on air quality.
